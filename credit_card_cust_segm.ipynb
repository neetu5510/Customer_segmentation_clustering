```python
inertia = []
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k, random_state=12).fit(df2_scaled2)
    inertia.append(np.sqrt(kmeans.inertia_))
```

```python
plt.figure(figsize=(18,6))
plt.title('optimal number of clusters')
plt.plot(range(1, 20), inertia, marker='s');
# from here we see optimal number of clusters is 4
```

```python
from sklearn import metrics
metrics.silhouette_score(df2_scaled2, km3.labels_)
```

```python
range1 = range(3,15)
scores = []
for k in range1:
    kmeans = KMeans(n_clusters=k, random_state=1).fit(df2_scaled2)
    scores.append(metrics.silhouette_score(df2_scaled2, kmeans.labels_))
```

```python
plt.figure(figsize=(15,5)) 
plt.plot(range1, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.grid(True)
```

```python
# finally from profiling and other methods we conclude that the optimal no. of clusters is 4
```

# Cluster description and market stategy

Cluster 0: These customers constitutes 20% of the total customer base.This is a risky group since they are mostly doing on off purchase.There is no installment purchase

Cluster 1: These are the potential target customers and constitutes approximately 25% of the customer base. The are taking cash, doing purchases on installments. We can encourage them for doing more purchases by providing suitable offers.

Cluster 2: These are the customers who are taking more cash advance and doing less purchase. We can target them by providing lower interest rates.They are 24% of the toal customer base.

Cluster 3: This group is doing all well.They are taking cash, doing purchases and paying on time. We can encourage them for 
buying more by providing lower interest rates. This group is 31% of the total customer base.
